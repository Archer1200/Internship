```python
cat_images = ["c1.jpg", "c2.jpg", "c3.jpg", "c4.jpg", "c5.jpg", "c6.jpg", "c7.jpg", "c8.jpg", "c9.jpg", "c10.jpg"]
dog_images = ["d1.jpg", "d2.jpg", "d3.jpg", "d4.jpg", "d5.jpg", "d6.jpg", "d7.jpg", "d8.jpg", "d9.jpg", "d10.jpg"]
eagle_images = ["e1.jpg", "e2.jpg", "e3.jpg", "e4.jpg", "e5.jpg", "e6.jpg", "e7.jpg", "e8.jpg", "e9.jpg", "e10.jpg"]
panther_images = ["p1.jpg", "p2.jpg", "p3.jpg", "p4.jpg", "p5.jpg", "p6.jpg", "p7.jpg", "p8.jpg", "p9.jpg", "p10.jpg"]
wolf_images = ["w1.jpg", "w2.jpg", "w3.jpg", "w4.jpg", "w5.jpg", "w6.jpg", "w7.jpg", "w8.jpg", "w9.jpg", "w10.jpg"]
```

```python
import cv2
```

```python
all_files = cat_images + dog_images + eagle_images + panther_images + wolf_images
```

```python
image_folder = "C:/Users/Himanshu/NEW PROJECT/facereco/"
```

```python
images = []
for name in all_files:
    img = cv2.imread(name)

    if img is not None:
        img = cv2.resize(img, (64, 64))  # Resize to 64x64
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
        images.append(img)
    else:
        print("Could not read:", name)
```

```python
images
```

```python
import numpy as np
images = np.array(images)
```

```python
labels = [0]*len(cat_images) + [1]*len(dog_images) + [2]*len(eagle_images) + [3]*len(panther_images) + [4]*len(wolf_images)
y = np.array(labels)
```

```python
x = images.reshape(len(images), -1)
```

```python
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(train_x, train_y)
```

```python
model.predict(test_x)
```

```python
model.score(test_x, test_y)
```
